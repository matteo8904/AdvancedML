```python
import sys
sys.path.insert(1, 'C:/Users/fabio/Desktop/ML4Fin Unibo')

import numpy  as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

import utils
```

<h3> notation and constants

>    setting constants to be used throughout the notebook

```python
seed = 1986436

rand = np.random.RandomState(seed)
mInt = (1 << 15)  # bitshift operator
MInt = (1 << 16)  # \equiv 2**16

log_n = 14
NUM = (1 << log_n)

EPS = 0.04

outputPrfx = "full"
challengePrfx = "challenge"
targetPrfx = "target"

bounds = { "T":      [6./12., 2.00]
         , "Sgma":   [ .15  ,  .80]
         , "Strike": [ .60  , 1.20]
         }
```

<h3> parameters generation

   > generating random parameters for the Black-Scholes model and contract specifications

```python
xDF = utils.lhs_sampling(rand, NUM, bounds=bounds)
utils.histo_params( xDF, title = "Random parameters density functions")
```

<h3> price generated options 

```python
yDF = utils.gen_BnS(NUM, xDF)
```

>    visualize the head of the dataset

```python
print(yDF.head(5))
```

<h3> divide the dataframe into train and test

```python
train_set, test_set = train_test_split(yDF, test_size=0.25, random_state=rand.randint(mInt,MInt))     # presence of a RandomState imposes a 
# random shaffling to be applied before the dataset is splitted (you can substitute by an int for reproducibility of the output)
```

<h3> add noise to prices in the train set

```python
if EPS > 0.0: 
    train_set_pnoise = utils.add_noise(rand, train_set, EPS)
else:         
    train_set_pnoise = train_set
```

>   plot noise vs true data

```python
fig, ax = plt.subplots(1,1, figsize=(12,6))

ax.plot(train_set["Price"], train_set_pnoise["Price"], ".", markersize=1)
ax.plot(train_set["Price"], train_set["Price"], color='red', linewidth=1)
ax.set_xlabel("(true) price")
ax.set_ylabel("noised price")
plt.show()
```

<h3> separate the test set into chalng (X) and target (Y)

```python
target = pd.DataFrame({"Price": test_set["Price"]})
chalng = test_set.drop(columns="Price")
```

<h3> write to disk

>    full train set (w or w/o noise)

```python
outputFile = "%s_%d.csv" %(outputPrfx, seed)
train_set_pnoise.to_csv(outputFile, sep=',', float_format="%.6f", index=False)
print("@ %-11s: training data frame written to '%s'" %("Info", outputFile))
print(train_set_pnoise)
```

>    challenge set

```python
challengeFile = "%s_%d.csv" %(challengePrfx, seed)
chalng.to_csv(challengeFile, sep=',', float_format="%.6f", index=False)
print("@ %-11s: challenge data frame written to '%s'" %("Info", challengeFile))
print(chalng)
```

>    target vector

```python
targetFile = "%s_%d.csv" %(targetPrfx, seed)
target.to_csv(targetFile, sep=',', float_format="%.6f", index=False)
print("@ %-11s: target data frame written to '%s'" %("Info", targetFile))
print(target)
```
